```python
#4. Append needed columns for crypto currency .
bitcoin_df['coin']=""


#ethereum_df
ethereum_df['coin']=""

#btccash_df
btccash_df['coin']=""

bitcoin_df
```

```python
#5. Iterate over the data frame row by row.
for index_label, row_series in bitcoin_df.iterrows():
   # For each row update the 'bitcoin_df' the name of the crypto currency is added.
   bitcoin_df.at[index_label , 'coin'] = str('Bitcoin')

for index_label, row_series in btccash_df.iterrows():
   # For each row update the 'btccash_df' the name of the crypto currency is added.
   btccash_df.at[index_label , 'coin'] = str('btccash')
    
for index_label, row_series in ethereum_df.iterrows():
   # For each row update the 'ethereum_df' the name of the crypto currency is added.
   ethereum_df.at[index_label , 'coin'] = str('ethereum')

ethereum_df
```

```python
#6. Drop not needed columns.
clean_bitcoin_df=bitcoin_df.drop(columns=['Unnamed: 0'])

clean_ethereum_df=ethereum_df.drop(columns=['Unnamed: 0'])

clean_btccas_df=btccash_df.drop(columns=['Unnamed: 0'])

clean_ethereum_df
```

```python
#7. Set index to the crypto name.
index_bitcoin_df=clean_bitcoin_df.set_index('coin')


index_ethereum_df=clean_ethereum_df.set_index('coin')

index_btccas_df=clean_btccas_df.set_index('coin')

index_ethereum_df
```

```python
be_df=index_bitcoin_df.append(index_ethereum_df, ignore_index=False)
final_df=be_df.append(index_btccas_df, ignore_index=False)
final_df.reset_index(inplace=True)
final_df
```

```python
final_df.to_csv('./final_df.csv', index=False)
```

```python
engine=sqlalchemy.create_engine('sqlite:///final_df', echo=True)
sqlite_connection = engine.connect()
```

```python
sqlite_table = "Crypto-1"
final_df.to_sql(sqlite_table, sqlite_connection, if_exists='fail')
```
